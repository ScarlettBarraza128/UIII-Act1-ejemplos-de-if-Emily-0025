```python
# 1. Ejemplo de if simple
# Este ejemplo utiliza operadores relacionales para verificar si un número está dentro de un rango específico.
numero = 15

if numero >= 10 and numero <= 20:
    print(f"El número {numero} está en el rango de 10 a 20.")
```

```python
# 2. Ejemplo de if-else
# Este ejemplo identifica si un año es bisiesto combinando operadores aritméticos, relacionales y lógicos.
año = 2024

if (año % 4 == 0 and año % 100 != 0) or (año % 400 == 0):
    print(f"{año} es un año bisiesto.")
else:
    print(f"{año} no es un año bisiesto.")
```

```python
# 3. Ejemplo de if-elif-else
# Este ejemplo clasifica un número introducido por el usuario y realiza un cálculo aritmético si es positivo.
numero = int(input("Ingresa un número: "))

if numero > 0:
    resultado = numero * 2
    print(f"El número es positivo. Su doble es {resultado}.")
elif numero < 0:
    print("El número es negativo.")
else:
    print("El número es cero.")
```

```python
# 4. Ejemplo con Operadores Lógicos y Aritméticos avanzados
# Este ejemplo determina el acceso a un descuento basado en la edad y el monto de una compra.
edad = 25
monto_compra = 150.0

if edad > 18 and (monto_compra - 50) > 100:
    print("Tienes acceso al descuento premium.")
else:
    print("No cumples los requisitos para el descuento.")
```
